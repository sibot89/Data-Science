```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
from google.colab import files
files.upload()
```

```python
raw_data = pd.read_csv('Binary Predictors.csv')
raw_data
```

```python
data = raw_data.copy()
data['Admitted'] = data['Admitted'].map({'Yes':1,'No':0})
data['Gender'] = data['Gender'].map({'Female':1,'Male':0})
data
```

Logit Regression with no independent variable

```python
y = data['Admitted']
x0 = np.ones(168)

reg_log0 = sm.Logit(y, x0)
results_log0 = reg_log0.fit()
results_log0.summary()
```

Adding one independent variable to the model

```python
y = data['Admitted']
x1 = data['Gender']

x = sm.add_constant(x1)

reg_log = sm.Logit(y, x)
results_log = reg_log.fit()
results_log.summary()
```

Adding variable 'SAT' to the model

```python
y = data['Admitted']
x1 = data[['SAT', 'Gender']]

x = sm.add_constant(x1)

reg_log = sm.Logit(y, x)
results_log = reg_log.fit()
results_log.summary()
```

```python
y = data['Admitted']
x1 = data[['SAT','Gender','Random']]

x = sm.add_constant(x1)

reg_log = sm.Logit(y, x)
results_log = reg_log.fit()
results_log.summary()
```

As you can see, the variable 'Random' doen't make the model better, so we will remove it.

```python
y = data['Admitted']
x1 = data[['SAT', 'Gender']]

x = sm.add_constant(x1)

reg_log = sm.Logit(y, x)
results_log = reg_log.fit()
results_log.summary()
```

```python
np.set_printoptions(formatter={'float': lambda x: "{0:0.2f}".format(x)})

results_log.predict()
```

```python
pr_table = results_log.pred_table(threshold=0.5)

accuracy_train = (pr_table[0,0] + pr_table[1,1])/pr_table.sum()
accuracy_train
```

```python
test = pd.read_csv('Test dataset.csv')
test
```

```python
test['Admitted'] = test['Admitted'].map({'Yes': 1, 'No': 0})
test['Gender'] = test['Gender'].map({'Female': 1, 'Male': 0})
test
```

```python
test_actual = test['Admitted']
test_data = test.drop(['Admitted'], axis=1)
test_data = sm.add_constant(test_data)
test_data
```

```python
def confusion_matrix(data, actual_values, model):
  pred_values = model.predict(data)
  bins = np.array([0, 0.5, 1])

  cm = np.histogram2d(actual_values, pred_values, bins=bins)[0]

  accuracy = round((cm[0,0] + cm[1,1])/cm.sum(), 4)

  return cm, accuracy
```

```python
cm = confusion_matrix(test_data, test_actual, results_log)
cm
```

```python
cm_df = pd.DataFrame(cm[0])

cm_df.columns = ['Predicted 0','Predicted 1']

cm_df = cm_df.rename(index={0: 'Actual 0',1:'Actual 1'})

cm_df
```
