```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
observations = 1000

xs = np.random.uniform(-10, 10, (observations,1))
ys = np.random.uniform(-10, 10, (observations,1))

inputs = np.column_stack((xs, ys))
inputs.shape
```

```python
noise = np.random.uniform(-1, 1, (observations,1))

targets = 4*xs - 5*ys + 9 + noise

print(targets.shape)
```

```python
targets = targets.reshape((observations,))
xr = xs.reshape((observations,))
yr = ys.reshape((observations,))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(xr, yr, targets)

ax.set_xlabel('xs')
ax.set_ylabel('ys')
ax.set_zlabel('Targets')

ax.view_init(azim=100)

plt.show()
```

```python
targets = targets.reshape(observations,1)
```

```python
init_range = 0.1

weights = np.random.uniform(-init_range, init_range, (2,1))

bias = np.random.uniform(-init_range, init_range, 1)

print(weights)
print(bias)
```

```python
learning_rate = 0.02
```

```python
for i in range(1000):
  outputs = np.dot(inputs, weights) + bias
  deltas = outputs - targets
  loss = np.sum(deltas ** 2) /2 /observations
  print(loss)

  deltas_scaled = deltas / observations

  weights = weights - learning_rate * np.dot(inputs.T, deltas_scaled)
  bias = bias - learning_rate * np.sum(deltas_scaled)
```

```python
print(weights)
print(bias)
```

```python
plt.plot(outputs, targets)

plt.xlabel('outputs')
plt.ylabel('targets')

plt.show()
```
